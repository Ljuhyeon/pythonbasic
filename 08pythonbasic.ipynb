#### -> 가변-위치 인자(var-positional parameters)

```python
def var_positional(*args): # 가변 변수로 사용할 변수 앞에 *를 하나 붙이면 가변-위치 인자로 사용할 수 있다.
    print(type(args))
    return sum([_ for _ in args])
```
