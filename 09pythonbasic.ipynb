# 예외(Exception)
------

#### 예외(Exception)란 코드를 실행하다가 에러가 발생하는 것을 의미합니다.


### 예외처리, try. except

```python
# 에러가 발생되는, 예외가 발생했을 때도 프로그램의 중단없이 계속 실행되도록 처리하는 것을 예외 처리라고 합니다.
# 이때는 실행할 코드를 try의 코드 블록에 적고 에러가 발생되어 예외 처리되는 부분을 except 코드 블록에 적습니다.
```

```python
try:
    # 나눗셈 연산 시 0으로 나누면 에러 발생
    result = 128/0
    print(f'연산의 결과는 {result}입니다.')
except:
    print('0으로 나눌 수 없습니다.')
```

```python
%%time
a = []
for _ in range(1000):
    a.append((_, _, _))
```

```python
try:
    x, y = input('숫자 두개를 입력하세요.').split(',')
except:
    print('값을 잘못 입력하셨습니다.')
```

### 예외 처리. except 에러마다 분기하기

```python
x = range(4)

try:
    for i in range(4):
        y = 10 / x[i]
except ZeroDivisionError as e:
    print('0으로 나눌 수 없습니다.')
    print(e)
except IndexError as e:
    print('인덱스 에러 발생')
    print(e)
```

### 예외 처리. else

```python
x = range(4)

try:
    for i in range(4):
        y = 10 / x[i]
except ZeroDivisionError as e:
    print('0으로 나눌 수 없습니다.')
    print(e)
except IndexError as e:
    print('인덱스 에러 발생')
    print(e)
else:
    print('에러없이 실행되었습니다.')
```

```python
%%time
try:
    x, y = map(int, input("값 입력:").split(' '))
    result = x / y
except ValueError:
    print('값을 잘못 입력하셨습니다.')
except ZeroDivisionError:
    print('0으로 나눌 수 없습니다.')
else:
    print(result)
```

### 예외 처리. finally

#### try와 finally를 사용하면 에러가 발생하더라도 무조건 실행된다.

```python
try:
    raise KeyError
finally:
    print('무조건 실행됩니다.')
```

```python
try:
    x, y = map(int, input("값 입력:").split(' '))
    result = x / y
except ValueError:
    print('값을 잘못 입력하셨습니다.')
except ZeroDivisionError:
    print('0으로 나눌 수 없습니다.')
else:
    print(result)
finally:
    print('프로그램을 종료합니다.')
```
