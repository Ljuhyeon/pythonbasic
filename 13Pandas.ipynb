## > Series에서 NaN이 아닌 값 구하기

```python
# 인구의 증가율을 NaN값을 배제하고 구해보기
rs = (s - s2) / s2 * 100
rs = rs[rs.notnull()] # rs.notnull은 boolean 값을 나타내고, 여기서 인덱싱을 해주면 True 값만 나온다
rs
```
