```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax1 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.title('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
# 
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns 
titanic = sns.load_dataset("titanic")
```

```python
bins = [0, 20, 30, 50, 70, 100]
labels = ["", "", "", "", ""]
explode = (0.1, 0, 0, 0, 0)
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral", "orange"]
C = pd.cut(titanic.age, bins, labels=labels).dropna()
P = titanic.value_counts(C, normalize=True)
plt.pie(list(P), explode=explode, labels=P.index, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90) 
plt.axis('equal')
plt.show()
```

```python
methods = {
    None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16', 'spline36',
    'hanning', 'hamming', 'hermite', 'kaiser', 'quadric', 'catrom', 'gaussian',
    'bessel', 'mitchell', 'sinc', 'lanczos'
}
fig, axes = plt.subplots(3, 6, figsize = (12, 6),
                         subplot_kw = {'xticks' : [], 'yticks' : []})
for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(X, cmap = plt.cm.bone_r, interpolation = interp_method)
    ax.set_title(interp_method)
    
plt.show()
```

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("3D Surface Plot")
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
```
